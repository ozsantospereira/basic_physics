# The physical model and equations

From the experimental setup, we define the following variables:  

- $ M_0 $ = Mass reading on the balance when the bar is outside the liquid.  
- $ M $ = Mass reading on the balance when the bar is partially immersed.  
- $ V_0 $ = Initial volume of the liquid before submersion.  
- $ V $ = New volume of the liquid after submersion.  
- $ V_{\text{displaced}} = V - V_0 $ = Volume of liquid displaced by the submerged part of the bar.  
- $ \rho_{\text{liquid}} $ = Density of the liquid.  
- $ g $ = Acceleration due to gravity.  
- $ E $ = Buoyant force acting on the bar.

To derive the correct relation between the measured mass $ M $ on the balance and the displaced volume $ V $, we apply **Archimedes' Principle** and Newton’s second law. From the experimental setup, we define the key variables: $ M_0 $ as the mass reading on the balance when the bar is outside the liquid, $ M $ as the mass reading when the bar is partially immersed, $ V_0 $ as the initial volume of the liquid before submersion, and $ V $ as the new volume after submersion. The displaced volume is given by 
$$ V_{\text{displaced}} = V - V_0 ,$$ 
which represents the portion of the liquid replaced by the submerged part of the bar. Additionally, we define $ \rho_{\text{liquid}} $ as the density of the liquid, $ g $ as the acceleration due to gravity, and $ E $ as the buoyant force acting on the bar.

By **Archimedes' Principle**, the buoyant force $ E $ is equal to the weight of the displaced liquid:
$$
E = \rho_{\text{liquid}} \cdot g \cdot V_{\text{displaced}}
$$
Since this force acts upward on the bar, its reaction force acts downward on the liquid, contributing to the measured weight on the balance. Before submersion, the balance reading corresponds to the force exerted by the container and the liquid:
$$
M_0 g = (\text{mass of liquid + container}) \cdot g
$$
After submersion, the balance also accounts for the reaction force due to buoyancy:
$$
M g = M_0 g + E
$$
Substituting 
$$
E = \rho_{\text{liquid}} g V_{\text{displaced}} 
$$ 
into the equation:
$$
M g = M_0 g + \rho_{\text{liquid}} g V_{\text{displaced}}
$$
Canceling $ g $ from both sides, we obtain:
$$
M = M_0 + \rho_{\text{liquid}} V_{\text{displaced}}
$$
Since $ V_{\text{displaced}} = V - V_0 $, we arrive at the final equation:
$$
M = M_0 + \rho_{\text{liquid}} (V - V_0)
$$
This equation establishes a **linear relationship** between $ M $ and $ V $. If we plot $ M $ against $ V $, the equation takes the form:
$$
M = \rho_{\text{liquid}} V + \left( M_0 - \rho_{\text{liquid}} V_0 \right)
$$
where the **slope** of the graph is $ \rho_{\text{liquid}} $, representing the density of the liquid, and the **intercept** is $ M_0 - \rho_{\text{liquid}} V_0 $, a constant offset. The density of the liquid can be directly determined from the slope of the linear fit.

# Fitting data with Minumum Least Squares

The **least squares method** determines the best-fit line for a dataset by minimizing the sum of squared residuals. Given a set of data points $(x_i, y_i)$, the goal is to determine the coefficients $ a $ (slope) and $ b $ (intercept) for the linear equation:

$$
y = ax + b
$$

**Step 1: Formulas for Least Squares Regression**  
The slope $ a $ and intercept $ b $ are given by:

$$
a = \frac{N \sum x_i y_i - \sum x_i \sum y_i}{N \sum x_i^2 - (\sum x_i)^2}
$$

$$
b = \frac{\sum y_i - a \sum x_i}{N}
$$

where:  
- $ N $ is the number of data points,  
- $ x_i, y_i $ are the observed data values.  

**Step 2: Estimating Errors**  
The standard errors of $ a $ and $ b $ are given by:

$
\sigma_a = \sigma_y \sqrt{\frac{1}{N \sum x_i^2 - (\sum x_i)^2}}
$

$
\sigma_b = \sigma_y \sqrt{\frac{\sum x_i^2}{N \sum x_i^2 - (\sum x_i)^2}}
$

where the residual standard error $ \sigma_y $ is defined as:

$
\sigma_y = \sqrt{\frac{\sum (y_i - (a x_i + b))^2}{N - 2}}
$

These formulas provide a measure of how well the regression line fits the data, with $ \sigma_a $ and $ \sigma_b $ indicating the uncertainty in the estimated slope and intercept, respectively.

# **Derivation of the Least Squares Regression Equations**

We aim to fit a **linear model** of the form:
$$
y = ax + b
$$
given a dataset of $ N $ points: $ (x_1, y_1), (x_2, y_2), \dots, (x_N, y_N) $. The goal is to determine the **best-fitting** values of $ a $ (slope) and $ b $ (intercept) using the **least squares method**.

**Step 1: Define the Error Function (Residual Sum of Squares)**  
The **error** (or residual) for each data point is:
$$
\varepsilon_i = y_i - (ax_i + b)
$$
The **least squares approach** minimizes the sum of squared errors:
$$
S = \sum_{i=1}^{N} \varepsilon_i^2 = \sum_{i=1}^{N} (y_i - ax_i - b)^2
$$
To find the best values of $ a $ and $ b $, we take the **partial derivatives** of $ S $ with respect to $ a $ and $ b $ and set them to zero.

**Step 2: Compute the Normal Equations**  
Partial derivative with respect to $ a $:
$$
\frac{\partial S}{\partial a} = \sum_{i=1}^{N} 2(y_i - ax_i - b)(-x_i) = 0
$$
Expanding:
$$
\sum_{i=1}^{N} x_i y_i - a \sum_{i=1}^{N} x_i^2 - b \sum_{i=1}^{N} x_i = 0
$$
Partial derivative with respect to $ b $:
$$
\frac{\partial S}{\partial b} = \sum_{i=1}^{N} 2(y_i - ax_i - b)(-1) = 0
$$
Expanding:
$$
\sum_{i=1}^{N} y_i - a \sum_{i=1}^{N} x_i - N b = 0
$$

**Step 3: Solve for $ a $ and $ b $**  
We now solve for $ a $ and $ b $. From the second equation:
$$
\sum y_i = a \sum x_i + N b
$$
Solving for $ b $:
$$
b = \frac{\sum y_i - a \sum x_i}{N}
$$
Substituting $ b $ into the first equation:  
$$
\sum x_i y_i - a \sum x_i^2 - \frac{\sum x_i (\sum y_i - a \sum x_i)}{N} = 0
$$
Multiplying everything by $ N $ to clear the fraction:
$$
N \sum x_i y_i - a N \sum x_i^2 - \sum x_i \sum y_i + a \sum x_i^2 = 0
$$
Rearranging:
$$
a (N \sum x_i^2 - (\sum x_i)^2) = N \sum x_i y_i - \sum x_i \sum y_i
$$
Finally, solving for $ a $:
$$
a = \frac{N \sum x_i y_i - \sum x_i \sum y_i}{N \sum x_i^2 - (\sum x_i)^2}
$$
Plugging $ a $ back into the equation for $ b $:
$$
b = \frac{\sum y_i - a \sum x_i}{N}
$$

**Step 4: Compute the Errors in $ a $ and $ b $**  
To compute the **uncertainties** (errors) in $ a $ and $ b $, we first determine the residual standard error $ \sigma_y $:
$$
\sigma_y = \sqrt{\frac{\sum (y_i - (ax_i + b))^2}{N - 2}}
$$
The standard errors of $ a $ and $ b $ are given by:
$$
\sigma_a = \sigma_y \sqrt{\frac{1}{N \sum x_i^2 - (\sum x_i)^2}}
$$
$$
\sigma_b = \sigma_y \sqrt{\frac{\sum x_i^2}{N \sum x_i^2 - (\sum x_i)^2}}
$$
These formulas give the uncertainties in our slope $ a $ and intercept $ b $, reflecting how much they may vary due to noise in the data.

**Final Summary of Equations**
$$
a = \frac{N \sum x_i y_i - \sum x_i \sum y_i}{N \sum x_i^2 - (\sum x_i)^2}
$$
$$
b = \frac{\sum y_i - a \sum x_i}{N}
$$
$$
\sigma_y = \sqrt{\frac{\sum (y_i - (ax_i + b))^2}{N - 2}}
$$
$$
\sigma_a = \sigma_y \sqrt{\frac{1}{N \sum x_i^2 - (\sum x_i)^2}}
$$
$$
\sigma_b = \sigma_y \sqrt{\frac{\sum x_i^2}{N \sum x_i^2 - (\sum x_i)^2}}
$$
These equations provide a complete least squares regression analysis, allowing us to estimate both the best-fit line and the uncertainties in its coefficients.

# Functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_experimental_data(n_points=10,M0=300,V0=100,rho=1,V_min=110,V_max=200,err_M=0.025,err_V=0.10,noise=5):
    # Generate volume values evenly spaced between V_min and V_max.
    V_values = np.linspace(V_min, V_max, n_points)
    # Calculate the mass values from the linear relation plus noise.
    M_values = rho * V_values + (M0 - rho * V0) + np.random.uniform(-noise, noise, n_points)
    # Create the DataFrame including the measurement uncertainties.
    df = pd.DataFrame({
        'M': M_values,
        'sigma_M': M_values * err_M,
        'V': V_values,
        'sigma_V': V_values * err_V     
    })
    return df

def calculate_fit(df, M0=300, V0=100, rho=1):
    # Extract the independent (x) and dependent (y) values.
    V_values = df['V'].values
    M_values = df['M'].values
    N = len(V_values)
    # Calculate necessary summations.
    sum_x = np.sum(V_values)
    sum_y = np.sum(M_values)
    sum_x2 = np.sum(V_values ** 2)
    sum_xy = np.sum(V_values * M_values)
    # Denominator for the slope and intercept formulas.
    D = N * sum_x2 - sum_x ** 2
    # Calculate slope (a) and intercept (b).
    a = (N * sum_xy - sum_x * sum_y) / D
    # 
    b = (sum_y - a * sum_x) / N
    # Predicted Mass values using the regression line.
    y_pred = a * V_values + b
    # Residual standard error.
    sigma_y = np.sqrt(np.sum((M_values - y_pred) ** 2) / (N - 2))
    # Compute Sxx for calculating uncertainties.
    Sxx = sum_x2 - (sum_x ** 2) / N
    #
    sigma_a = sigma_y / np.sqrt(Sxx)
    # Standard error for the intercept.
    x_bar = sum_x / N
    #
    sigma_b = sigma_y * np.sqrt(1 / N + (x_bar ** 2) / Sxx)
    #
    fit_params = {
        'M0': M0,
        'V0': V0,
        'rho': rho,
        'a': a,
        'sigma_a': sigma_a,
        'b': b,
        'sigma_b': sigma_b,
        'y_pred': y_pred,
        'sigma_y': sigma_y,
    }
    tb = pd.DataFrame([fit_params])
    return tb, fit_params
    

def plot_regression(df, fit_results):
    V_values = df['V'].values
    M_values = df['M'].values
    y_pred = fit_results['y_pred']
    a = fit_results['a']
    b = fit_results['b']
    sigma_a = fit_results['sigma_a']
    sigma_b = fit_results['sigma_b']
    plt.figure(figsize=(8, 5))
    # Plot experimental data as scatter points.
    plt.scatter(V_values, M_values, color='r', label="Experimental Data")
    # Plot the regression line.
    plt.plot(V_values, y_pred, linestyle='--', color='b',
             label=f"Best Fit: y = ({a:.2f} ± {sigma_a:.2f})x + ({b:.2f} ± {sigma_b:.2f})")
    plt.xlabel("Volume (mL)")
    plt.ylabel("Mass (g)")
    plt.title("Least Squares Regression: Mass vs. Volume")
    plt.grid(True)
    plt.legend()
    plt.show()
    
def format_table(df):
    return pd.DataFrame({
        "$(M_i \\pm \\delta M_i) \\ g$": [f"{M:.3f} ± {sigma_M:.3f}" for M, sigma_M in zip(df["M"], df["sigma_M"])],
        "$\\frac{\\delta M_i}{M_i}$": [f"{(sigma_M / M):.3f}" for M, sigma_M in zip(df["M"], df["sigma_M"])],
        "$(V_i \\pm \\delta V_i) \\ ml$": [f"{V:.3f} ± {sigma_V:.3f}" for V, sigma_V in zip(df["V"], df["sigma_V"])],
        "$\\frac{\\delta V_i}{V_i}$": [f"{(sigma_V / V):.3f}" for V, sigma_V in zip(df["V"], df["sigma_V"])]
    })
```

# Analysis

```python
# 1. Generate the experimental data (with measurement uncertainties).
df = generate_experimental_data(n_points=10,M0=200,V0=100,rho=1,V_min=110,V_max=200,err_M=0.025,err_V=0.050,noise=3)

# 2. Calculate the linear fit parameters.
tb1, fit_params = calculate_fit(df, M0=300, V0=100, rho=1)

# Creates a formated table for the experiment
tb2 = format_table(df)
```
